```python
import pandas as pd

df = pd.read_csv("n_movies.csv")


def classify_type(row):
    year = str(row["year"])
    cert = str(row["certificate"])

    if "–" in year or cert.startswith("TV-"):
        return "Serie"
    else:
        return "Pelicula"


df["Tipo"] = df.apply(classify_type, axis=1)


df = df.drop(columns=["certificate"])



df.head(100)
```

```python

columnas_es = {
    "title": "Titulo",
    "year": "Año",
    "genre": "Genero",
    "duration": "Duracion",
    "rating": "Clasificacion",
    "description": "Descripcion",
    "stars": "Estrellas",
}
df2 = df2.rename(columns=columnas_es)

df2
```

```python
duplicados = df2[df2["Titulo"].duplicated(keep=False)]


duplicados = duplicados.sort_values(by="Titulo")

duplicados

```

```python

df2 = (
    df2.sort_values(by=["Titulo", "Clasificacion"], ascending=[True, False])
      .drop_duplicates(subset="Titulo", keep="first")
)
```

```python
df2["Titulo"].duplicated().sum()
```

```python


df2["duracion_num"] = (
    df2["Duracion"]
    .str.replace(" min", "", regex=False)
    .astype(float)
)


df2["genero_principal"] = df2["Genero"].str.split(",").str[0]

media_por_genero = (
    df2[
        (df2["Tipo"] == "Pelicula") &
        (df2["duracion_num"].notnull())
    ]
    .groupby("genero_principal")["duracion_num"]
    .mean()
)

media_global_peliculas = media_por_genero.mean()

def imputar_duracion(row):
    if row["Tipo"] == "Pelicula" and pd.isnull(row["Duracion"]):
        genero = row["genero_principal"]
        if genero in media_por_genero:
            return f"{round(media_por_genero[genero])} min"
        else:
            return f"{round(media_global_peliculas)} min"

    if row["Tipo"] == "Serie" and pd.isnull(row["Duracion"]):
        return "45 min"

    return row["Duracion"]

df2["Duracion"] = df2.apply(imputar_duracion, axis=1)


df2["duration_min"] = (
    df2["Duracion"]
    .str.replace(" min", "", regex=False)
    .astype(float)
)
df2
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import urllib.parse
import time
from tqdm import tqdm


INPUT_CSV = "movies_final.csv"
OUTPUT_CSV = "movies_with_images.csv"

TMDB_API_KEY = "6ce29e5b1926e1456f60ce528aaa9f51"

PLACEHOLDER_IMAGE = "https://upload.wikimedia.org/wikipedia/commons/a/ac/No_image_available.svg"

HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "es-ES,es;q=0.9,en;q=0.8"
}

TMDB_IMAGE_BASE = "https://image.tmdb.org/t/p/w500"

# Wikipedia 
def wikipedia_image(search_term):
    try:
        query = urllib.parse.quote(search_term)
        url = f"https://en.wikipedia.org/wiki/{query}"

        r = requests.get(url, headers=HEADERS, timeout=5)
        if r.status_code != 200:
            return None

        soup = BeautifulSoup(r.text, "html.parser")
        infobox = soup.find("table", class_="infobox")

        if not infobox:
            return None

        img = infobox.find("img")
        if not img or not img.get("src"):
            return None

        src = img["src"].lower()

        if src.endswith(".svg") or "no_image_available" in src:
            return None

        return "https:" + img["src"]

    except:
        return None

# TMDB 
def tmdb_image(search_term):
    try:
        search_url = "https://api.themoviedb.org/3/search/movie"
        params = {
            "api_key": TMDB_API_KEY,
            "query": search_term,
            "language": "es-ES"
        }

        r = requests.get(search_url, params=params, timeout=5)
        if r.status_code != 200:
            return None

        data = r.json()
        if not data["results"]:
            return None

        movie = data["results"][0]
        poster_path = movie.get("poster_path")

        if not poster_path:
            return None

        return TMDB_IMAGE_BASE + poster_path

    except:
        return None



df = pd.read_csv(INPUT_CSV)

if "link" not in df.columns:
    df["link"] = ""

# Buscar imágenes
titles = df["Titulo"].unique()

print(f"🎬 Buscando imágenes de {len(titles)} películas...")

for title in tqdm(titles):
    img = wikipedia_image(title)

    if img is None:
        img = tmdb_image(title)

    if img is None:
        img = PLACEHOLDER_IMAGE

    df.loc[df["Titulo"] == title, "link"] = img

    time.sleep(0.25)

# -----------------------------
# Guardar CSV
# -----------------------------
df.to_csv(OUTPUT_CSV, index=False)
print("✅ CSV generado correctamente:", OUTPUT_CSV)
```

```python
df = df_raw.copy()

# Guardar original (por si quieres revisar)
df["Clasificacion_raw"] = df["Clasificacion"].astype(str)

# Normalizar coma decimal -> punto
df["Clasificacion_raw"] = df["Clasificacion_raw"].str.replace(",", ".", regex=False)

# Extraer el primer número tipo 7.8 / 8 / 6.5
df["Clasificacion_num"] = df["Clasificacion_raw"].str.extract(r"(\d+(?:\.\d+)?)")[0]

# Convertir a float
df["Clasificacion_num"] = pd.to_numeric(df["Clasificacion_num"], errors="coerce")

print("Total títulos:", len(df))
print("Con rating válido:", df["Clasificacion_num"].notna().sum())
print("Sin rating (NaN):", df["Clasificacion_num"].isna().sum())

df[["Clasificacion", "Clasificacion_raw", "Clasificacion_num"]].sample(10, random_state=1)

```

```python
# Solo quitamos filas donde falte lo básico
df = df.dropna(subset=["Clasificacion_num", "Genero", "Tipo"]).copy()
print("Después de filtrar (rating+genero+tipo):", len(df))

df[["Titulo","Genero","Tipo","Clasificacion_num"]].head()
```

```python
HIGH_TH = 7.5  # umbral de "alta" (cámbialo a 7.0 u 8.0 si quieres)

# Bins de valoración
df["rating_bin"] = pd.cut(
    df["Clasificacion_num"],
    bins=[-np.inf, 6.0, HIGH_TH, np.inf],
    labels=["Baja (<6.0)", f"Media (6.0-{HIGH_TH})", f"Alta (≥{HIGH_TH})"]
)

# Separar géneros (multi-género) y explotar a filas individuales
df_gen = df.copy()
df_gen["genre_single"] = df_gen["Genero"].astype(str).str.replace(r"\s+", " ", regex=True).str.strip().str.split(", ")
df_gen = df_gen.explode("genre_single")
df_gen["genre_single"] = df_gen["genre_single"].str.strip()

print("Filas tras explode (sube porque un título puede tener varios géneros):", len(df_gen))
df_gen[["Titulo","Genero","genre_single","Clasificacion_num","Tipo","rating_bin"]].head(10)
```
